```python
#set the index as a release_date 

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
df_tracks = pd.read_csv('tracks.csv')
df_tracks.set_index( 'release_date', inplace=True)
df_tracks.index= pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
#check the artist in the 18 row in a data set 
df_tracks[["artists"]].iloc[18]
```

```python
#convert duration ms in sec
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round (x/1000))
df_tracks.drop("duration_ms", inplace= True ,axis = 1)
df_tracks.duration.head()
```

```python
#corelation map 
corr_df= df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap= sns.heatmap(corr_df,annot= True, fmt=".1g", vmin = -1 , vmax = 2,center =0,cmap ="cool",linewidths=2,linecolor="Blue")
heatmap.set_title("correlation Heat map ")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation =45)
heatmap.set_yticklabels(heatmap.get_yticklabels(),rotation =0)
```

```python
#distribution plot to visulase the total num of song in each year  that is available on spotify
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
df_tracks = pd.read_csv('tracks.csv')
df_tracks.set_index( 'release_date', inplace=True)
df_tracks.index= pd.to_datetime(df_tracks.index)

df_tracks["dates"] =df_tracks.index.get_level_values("release_date")
df_tracks.dates =pd.to_datetime(df_tracks.dates) 
years=df_tracks.dates.dt.year            
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of song per year")
```

```python
#barplot
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round (x/1000))
total_dr=df_tracks.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_dr, ax=ax).set(title="year VS Duration")
plt.xticks(rotation = 90 )
```

```python
total_dr=df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_dr, ax=ax).set(title="year VS Duration")
plt.xticks(rotation =60 )
```
